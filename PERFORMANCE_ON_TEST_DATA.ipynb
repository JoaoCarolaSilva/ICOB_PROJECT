### Getting last elite median individuals from each model

#### make median individual from standard GA

```python
folder_path_standard = parent_folder_ICOB_project + 'GA\\logs_with_best_params\\'
```

```python
# Get all .txt file paths in the folder
file_paths = glob.glob(os.path.join(folder_path_standard, "*.txt"))

# Load each file into a DataFrame and store in a list
dataframes = [pd.read_csv(file, sep=",") for file in file_paths] 
```

```python
### get genotypes for last elites of each run 
standard_GA_last_elites = []

for run in dataframes:
    best_fitness_last = run.iloc[-1]['best_genotype']
    standard_GA_last_elites.append(best_fitness_last)
```

```python
### turn genotypes into lists (from string)
standard_GA_last_elites_ = [] # list of lists to save 
 
for genotype_string in standard_GA_last_elites:
    genotype_list = ast.literal_eval(genotype_string)
    standard_GA_last_elites_.append(genotype_list)
```

```python
### making the standard GA median final elite

# convert to numpy array and compute median along axis 0 (column-wise)
median_coefficients = np.median(np.array(standard_GA_last_elites_), axis=0)

# convert back to list if needed
median_indi_standardGA = median_coefficients.tolist()
```

### Seeing RMSE on test data for each median individual/model

```python
test_data_path = parent_folder_ICOB_project + 'parkinsons_CIFOBIO_TEST.csv'
```

```python
testdata = pd.read_csv(test_data_path)
```

```python
def calculate_fitness_RMSE(individual, data):
    '''
    This function calculates RMSE (measure of fitness) of an
    individual by receiving the individual (as list of coefficients
    where the 1st coefficient is the intercept) as well as the data
    '''
    # Split features and target
    X = data.iloc[:, :-1].values  # all columns except last
    y_true = data.iloc[:, -1].values  # last column

    # Separate intercept and coefficients
    intercept = individual[0]
    coefficients = np.array(individual[1:])

    # Compute predictions: y_pred = intercept + X @ coefficients
    y_pred = intercept + np.dot(X, coefficients)

    # Compute RMSE
    rmse = np.sqrt(np.mean((y_true - y_pred) ** 2))

    # Return RMSE as fitness (lower is better)
    return rmse
```

#### RMSE on test data standard GA

```python
calculate_fitness_RMSE(median_indi_standardGA, testdata)
```

#### RMSE on test data egalitarian GA

```python
calculate_fitness_RMSE(median_indi_egalitarianGA, testdata)
```

#### RMSE on test data elite GA

```python
calculate_fitness_RMSE(median_indi_eliteGA, testdata)
```

#### RMSE on test data weirdo GA

```python
calculate_fitness_RMSE(median_indi_weirdoGA, testdata)
```

#### RMSE on test data immigrant GA

```python
calculate_fitness_RMSE(median_indi_immigrantGA, testdata)
```
